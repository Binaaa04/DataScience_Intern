```python
import pandas as pd
import hvplot.pandas
import panel as pn

pn.extension('bokeh','tabulator')
df = pd.read_csv('/content/drive/MyDrive/MAGANG/DATA SCIENCE/mpg.csv')
df2 = df.dropna()
# Distribusi nilai mpg seluruh mobil
chart1 = df2.hvplot.hist(
    y='mpg',
    bins=20,
    title='Distribusi Nilai MPG (Miles per Gallon)',
    xlabel='MPG',
    ylabel='Jumlah Mobil'
)
# Perbedaan rata-rata mpg berdasarkan jumlah cylinders
chart2 = df2.groupby('cylinders')['mpg'].mean().hvplot.bar(
    title='Rata-rata MPG berdasarkan Jumlah Silinder',
    xlabel='Jumlah Silinder',
    ylabel='Rata-rata MPG'
)
# Hubungan antara weight dan mpg
chart3 = df2.hvplot.scatter(
    x='weight',
    y='mpg',
    title='Hubungan antara Berat Mobil dan Efisiensi (MPG)',
    xlabel='Berat Mobil',
    ylabel='MPG'
)

origin_filter = pn.widgets.Select(name='Pilih Origin', options=sorted(df2['origin'].unique()))
year_filter = pn.widgets.IntSlider(name='Pilih Tahun', start=int(df2['model_year'].min()), end=int(df2['model_year'].max()))

@pn.depends(origin_filter, year_filter)
def update_dashboard(origin, year):
    dff = df2[(df2['origin'] == origin) & (df2['model_year'] == year)]
    if len(dff) == 0:
        return pn.pane.Markdown("### Tidak ada data")

    hist = dff.hvplot.hist(y='mpg', bins=10, title='Distribusi MPG')
    scatter = dff.hvplot.scatter(x='weight', y='mpg', title='Hubungan Weight vs MPG')
    bar = dff.groupby('cylinders')['mpg'].mean().hvplot.bar(title='Rata-rata MPG per Cylinders')

    return (hist + scatter + bar).cols(2)

dashboard = pn.Column(
    "#Dashboard Analisis MPG",
    "Gunakan filter di bawah untuk melihat pola data mobil:",
    pn.Row(origin_filter, year_filter),
    update_dashboard
)

dashboard
```

**Insight dari Analisis:**

Mobil dengan jumlah silinder lebih sedikit umumnya memiliki MPG lebih tinggi (lebih irit).

Mobil berat cenderung memiliki MPG lebih rendah, menunjukkan hubungan negatif antara berat dan efisiensi bahan bakar.

Berdasarkan filter origin, mobil asal Jepang biasanya menunjukkan efisiensi bahan bakar lebih baik dibandingkan mobil dari Amerika.
