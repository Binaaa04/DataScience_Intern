```python
import matplotlib.pyplot as plt
import seaborn as sns

# === Load dataset ===
df = pd.read_csv('/content/drive/MyDrive/MAGANG/DATA SCIENCE/dataSalesCafeClean.csv', low_memory=False)

# Hitung jumlah transaksi per produk
produk_populer = df['Item'].value_counts().head(10)

plt.figure(figsize=(8,5))
sns.barplot(x=produk_populer.values, y=produk_populer.index, palette='viridis')
plt.title("Juice adalah Produk dengan Pembelian Tertinggi", fontsize=14, weight='bold')
plt.xlabel("Jumlah Transaksi")
plt.ylabel("Nama Item")
plt.show()
```

Berdasarkan visualisasi di atas, terlihat bahwa Juice merupakan item dengan jumlah transaksi tertinggi, mengungguli produk lain seperti Coffee, Salad, dan Cake. Ini menunjukkan bahwa minuman, khususnya Juice, menjadi pilihan utama konsumen dalam transaksi di kedai kopi ini.

Dari sisi business insight, tren ini mengindikasikan adanya preferensi pelanggan terhadap produk yang lebih segar atau sehat, yang bisa menjadi peluang untuk:

Mengembangkan variasi juice baru (misalnya seasonal flavors)

Menerapkan strategi bundling: misalnya Juice + makanan ringan (Cake atau Cookie) dengan harga promo

Menempatkan Juice sebagai "anchor product" di strategi pemasaran atau layout menu

Selain itu, item seperti Sandwich, Smoothie, dan Cookie juga menunjukkan performa cukup tinggi, sementara keberadaan kategori UNKNOWN dan ERROR menandakan potensi perbaikan dalam kualitas data input atau labeling produk di sistem POS.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# === Load dataset ===
df = pd.read_csv('/content/drive/MyDrive/MAGANG/DATA SCIENCE/dataSalesCafeClean.csv', low_memory=False)

pembayaran = df['Payment Method'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(pembayaran.values, labels=pembayaran.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Set2'))
plt.title("E-Wallet Mendominasi Metode Pembayaran", fontsize=14, weight='bold')
plt.show()
```

Berdasarkan hasil visualisasi, e-wallet menjadi metode pembayaran yang paling dominan dengan porsi 48,7%. Hal ini menunjukkan bahwa perilaku konsumen semakin bergeser ke arah pembayaran digital. Bisnis perlu memanfaatkan tren ini dengan memperluas kerja sama dengan penyedia e-wallet serta memastikan sistem pembayaran digital berjalan optimal, tanpa meninggalkan opsi pembayaran konvensional bagi segmen tertentu.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# === Load dataset ===
df = pd.read_csv('/content/drive/MyDrive/MAGANG/DATA SCIENCE/dataSalesCafeClean.csv', low_memory=False)

df['Transaction Date'] = pd.to_datetime(df['Transaction Date'])

# Hitung jumlah transaksi per hari
transaksi_harian = df.groupby('Transaction Date').size()

plt.figure(figsize=(10,5))
plt.plot(transaksi_harian.index, transaksi_harian.values, marker='o')
plt.title("Transaksi Tertinggi Terjadi pada 2 Juli 2023", fontsize=14, weight='bold')
plt.xlabel("Bulan")
plt.ylabel("Jumlah Transaksi")
plt.grid(alpha=0.3)
plt.show()
```

Berdasarkan visualisasi di atas, jumlah transaksi harian selama tahun 2023 cenderung stabil dengan rata-rata di kisaran puluhan transaksi per hari. Namun, terjadi lonjakan sangat tajam pada tanggal 2 Juli 2023 yang mencapai hampir 500 transaksi. Lonjakan ini menunjukkan adanya aktivitas khusus seperti promo besar, event penjualan, atau kerja sama strategis yang berdampak signifikan terhadap peningkatan transaksi. Dari sudut pandang bisnis, momen seperti ini perlu dianalisis lebih lanjut untuk dioptimalkan kembali pada periode berikutnya karena terbukti efektif dalam meningkatkan performa penjualan secara drastis.
